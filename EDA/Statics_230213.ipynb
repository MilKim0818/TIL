```python
liar = pd.read_excel('liar.xlsx')
liar.head()
# Creativity: 창의성점수(연속)
# Position: 등급(서열-순서가 있는 범주형)
# Novice: 초보자(1: 처음 나온 사람, 0: 경험자)
```

- Position별로 Creativity의 평균이 다른가? (여러 집단의 평균이 다른가? -> 분산분석)
- Novice와 아닌 사람 사이에 Creativity가 다른가?
- Novice와 아닌 사람 사이에 Position이 다른가?

```python
liar.Position.unique()
```

```python
#등분산성검증 
pg.homoscedasticity(dv='Creativity', group='Position',
                    data=liar)
```

```python
#분산이 다르다를 기각을 못하니까 그냥 anova를 한다.
pg.anova(dv='Creativity', between='Position', data=liar)
#귀무가설 : 모든 포지션의 평균 차이가 같다.
#귀무가설 기각 : 어떤 포지션은 평균이 다르다.
```

```python
pg.pairwise_tukey(dv='Creativity', between='Position', 
                  data=liar)
```

```python
#welch_anova는 분산이 다를때 해야하나봄ㅠㅠ
pg.welch_anova(dv='Creativity', between='Position', data=liar)
```

```python
p=0.007679
p<0.05
```

```python
#Novice가 0인 경우와 1인 경우의 Creativity의 평균 
#귀무가설 : Novice와 아닌 사람의 Creativity의 평균이 같다.
#0인 경우의 Creativity가 더 높다.
pg.ttest(liar[liar.Novice == 0].Creativity,
         liar[liar.Novice == 1].Creativity)
```
