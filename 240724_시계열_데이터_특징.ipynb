## 시계열 데이터 분석, 예측에 대한 기본 구조
* 접근 방법
  - 과거의 패턴이 시간의 흐름에 따라 반복될 것을 가정함

* 특징
 * 4요소  
1) 추세 (trend)
    * 장기적인 시계열 움직임 ex) 사람이 건강해짐에 따라 운동 평균 심박수가 감소하는 것.  

    2) 주기성 (cycle)  
    * 불규칙성보다 지속 기간이 길지만 추세를 구성하는 것보다 짧은 반복적인 변동이 발생  
    
    3) 계절성 (seasonal)
    * 1년보다 짧은 기간 동안 정기적으로 발생하는 현상(예: 봄과 가을에는 장거리 산책을 하기에 너무 춥지도 더우지도 않아 걸음 수가 더 많음)  
    
    4) 불규칙성 (Residual)
    * 일정한 규칙성을 인지할 수 없는 변동의 유형 (원래 데이터에서 추세, 주기성, 계절성, 불규칙성 뺀 나머지)
  
  * 시계열 데이터 분해 방법
  1) 덧셈 분해 (Additive)
  2) 곱셈 분해 (Multiplicative)
    

* 예측 방법
(시계열 데이터가 정성적/비정성적인지에 따라 예측 모델링 방법이 달라지는 것으로 보임..!)  
    1) 정상성(Stationarity) 시계열 데이터  
    * 정상적 = 데이터의 통계적 특성이 시간에 따라 변하지 않는다는 것
    * 통계적 특성 ex) 평균, 분산, 자기상관 등
    * 정상성을 판단하는 통계적 기법 있음.

    2) 비정상성 시계열 데이터
    * 비정상성 데이터의 경우, 정상성을 확보하기 위한 다양한 기법을 활용하여 예측 모델의 일관성과 정확성을 확보해야 함. ex) 차분, 로그 변환, 이동 평균 등
    * 보통 비정상성 시계열 데이터가 더 많은듯...

## 시계열 데이터 예시
* AirPassengers (시계열 데이터 분석 기본 예시로 많이 사용하는 데이터셋)
* 데이터 출처  
https://www.kaggle.com/datasets/rakannimer/air-passengers?resource=download


```python
# 데이터 가공
import pandas as pd
import numpy as np
# 시각화
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
csv_path = 'https://raw.githubusercontent.com/rahulhegde99/Time-Series-Analysis-and-Forecasting-of-Air-Passengers/master/airpassengers.csv'
df = pd.read_csv(csv_path)
print(df.shape)
df.head()
```

```python
# 시간 타입으로 변경하기
df['Month'] = pd.to_datetime(df['Month'])
df.set_index('Month', inplace=True)

# DatetimeIndex의 freq 설정
df = df.asfreq('MS')
df.head()
```

```python
# 시간 간격 (1달 단위)
print(df.index.nunique())
df.index.unique()
```

```python
df.groupby('Month').count()
```

```python
# 기술통계량
df.describe()
```

```python
# x축 시간, y측 관측값 (승객수) 그래프
# 0으로 뚝 떨어지는 구간이 없음 (결측값 없음)
df.plot(y='#Passengers', figsize=(10,5))
plt.show()
```

## 데이터의 실제 분포, 추세(경향성), 계절성, 잔차를 한번에 확인
* 파이썬 라이브러리 : `statsmodels`
* 함수 설명 : https://www.statsmodels.org/stable/generated/statsmodels.tsa.seasonal.seasonal_decompose.html#statsmodels.tsa.seasonal.seasonal_decompose



* R코드 예시 자료
image.png
https://rpubs.com/emb90/137525

```python
# 시계열 데이터의 3요소를 분해하기 !
# 패키지 설치
!pip install statsmodels
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Multiplicative 시계열 데이터 분해
result = seasonal_decompose(df['#Passengers'], model='multiplicative')
```

```python
# 결과 시각화
plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(df['#Passengers'], label='Observed')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(result.trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(result.seasonal, label='Seasonal')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(result.resid, label='Residual')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```
